```python
class SmartLight:
    def __init__(self, name):
        self.light_name = name
        self.is_on = False

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def display_status(self):
        if self.is_on:
            print(self.light_name, "is ON")
        else:
            print(self.light_name, "is OFF")


name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")

light = SmartLight(name)

if action.upper() == "ON":
    light.turn_on()
else:
    light.turn_off()

light.display_status()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

emp = Employee(name, emp_id, dept)
emp.display_id_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.display_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


name = input("Product Name: ")
price = int(input("Price: "))

product = Product(name, price)
product.display_price_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


name = input("Movie Name: ")
rating = float(input("Rating: "))

movie = Movie(name, rating)
movie.display_movie()
```

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


customer = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(customer, address)
delivery.display_details()
```
