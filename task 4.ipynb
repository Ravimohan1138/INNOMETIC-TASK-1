```python
def process_result(marks):
  total = 0

  for mark in marks:
    total += mark

  average = total / len(marks)

  if average >= 50:
    return "Pass"
  else:
    return "Fail"

marks = [60, 55, 40, 70]
result = process_result(marks)

print("Result:", result)
```

```python
def check_attendance(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance)

print("Attendance Status:", status)
```

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check minimum length
    if len(password) < 8:
        return "Weak Password"

    # Check for digit and special character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    # Final strength check
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter password: ")
result = check_password(password)
print(result)
```
